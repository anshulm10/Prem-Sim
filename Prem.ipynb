```python
def predictor(home, away):
  ahg = 1.51 #Avg Home Goals Scored
  ahc = 1.31 #Avg Home Goals Conceded
  aag = 1.31 #Avg Away Goals Scored
  aac = 1.51 #Avg Away Goals Conceded

  # Reciving the req Values
  if home in pred.index and away in pred.index:
    global homet, awayt
    homet = pred._get_value(home, 'Team')
    awayt = pred._get_value(away, 'Team')
    print(homet, "VS", awayt)
    lambdH = max(0, ahg + pred.at[home, "Home Attack"] - pred.at[away, "Away Defense"])
    lambdA = max(0, ahc + pred.at[away, "Away Attack"] - pred.at[home, "Home Defense"])
    
    print(homet, ":" ,lambdH, awayt, ":" , lambdA)
  
    Hw, Aw, D = 0, 0, 0 
    #Poisson Distribution
    #X and Y reprisents number of goals a team wiil score
    for x in range(1, 18): 
      for Y in range(1, 18):
        p = poisson.pmf(x, lambdH) * poisson.pmf(Y, lambdA)
        if lambdH == 0:
          p = lambdA
          Aw = p
          break
        elif lambdA == 0:
          p = lambdH
          Hw = p
          break
        else:
          if x == Y:
            D += p
          elif x > Y:
            Hw += p
            
          else:
            Aw += p

    print("Probablity of Home Win = ", Hw)
    print("Probablity of Away Win = ", Aw)
    print("Probablity of a Draw = ", D)    
    home, away = 0, 0    
    if Hw > Aw > D:
      home += 3
      print(homet ,home, "\n")
    elif Aw > Hw > D :
      away += 3
      print(awayt ,away, "\n")
    elif D > Hw > Aw :
      home += 1
      away += 1
      print(homet, home, awayt, away, "\n")
    elif D > Aw > Hw :
      home += 1
      away += 1
      print(homet, home, awayt, away, "\n")
    elif Hw > Aw:
      home += 3
      print(homet ,home, "\n")
    elif Aw > Hw:
      away += 3
      print(awayt ,away, "\n")
    else:
      home += 1
      away += 1
      print(homet, home, awayt, away, "\n")


for  h in range(1, 18):
  for a in range(1, 18):
    if h != a:
      predictor(h, a)
    
   
    
```
